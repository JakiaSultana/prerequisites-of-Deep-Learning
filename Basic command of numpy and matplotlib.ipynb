### Generate random Number From Array
#### choice() method allows us to generate a random value from an array

```python
from numpy import random
x = random.choice([3,5,7,9])
print("choice() takes an array and randomly returns one of the values from the array and that is: ",x)

x = random.choice([1,4,7,9], size =3)
print(x) #1D array

x = random.choice([1,6,3,9,10], size = (3,5))
print(x) # 3 arrays with 5 elements
```

### Random Data Distribution
#### Data Distribution is a list of all possible values, and how often each value occurs.
#### A Random distribution is a set of random numbers that follow a certail probability density function.
#### Probability Density Function describes a continuous probabilty
#### choice() method allows us to specify the probability for each value.

```python
from numpy import random
x = random.choice([1,5,6,8], p=[0.3,0.1,0.6,0.0], size =(3,5))
print(x) # means 8 has 0 probabily to occur
```

### Random Permutations

### Normal(Gaussian) Distribution
The curve of a Normal Distribution is known as Bell Curve because it is bell-shaped curve.

#### random.normal() mathod has 3 parameters: 
 Loc -(Mean) where the peak of the bell exists.
 
Scale - (Standard Deviation) how flat the graph distribution should be.

Size - The shape of the returned array.


```python
from numpy import random
x = random.normal(size=(2,3)) #size defines 2rows ans each contains 3 elements
print(x)

x = random.normal(loc=1, scale=2, size=(2,3))
print(x)
```

#### Visualization of Normal Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist = False)
plt.show()

sns.distplot(random.normal(170,10,250))
plt.show()
```

### Binomial Distribution
Binomial Distribution is a Discreate Distribution. It has 3 parameters:

n - number of trails

p - probability of occurance of each trail

size -  the shape of returned array

```python
#Given 10 trials for coin toss generate 10 data points:
from numpy import random
x = random.binomial(n=15, p= 0.5, size = 20)# size define the total num of output
print(x) #n defines the number from 0-15
```

### Visualization of Binomial Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=10, p = 0.5, size = 1000), hist = True, kde = False)
#kde parameter detarmines curve
plt.show()
```

### Difference Between Normal and Binomial Distribution
The main difference between normal and binomial distribution is that normal distribution is continuous whereas binomial is discreate. 

```python
sns.distplot(random.normal(loc= 50, scale = 5, size = 1000), hist = False, label ="normal")
sns.distplot(random.binomial(n= 100, p=0.5, size= 1000), hist= False, label = 'binomial')

plt.show()
```

### Poisson Distribution
It estimates how many times an event can happen in a specified time. e.g. If someone eats twice a day what is probability he will eat thrice?
#### It has 2 parameters:
lam - rate or known number of occurance

size - The shape of the returned array

```python
from numpy import random
x = random.poisson(lam = 2, size= 10)
print(x)
```

#### Visualization of Poisson Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam= 2, size = 1000), kde = False)
plt.show()
```

### Uniform Distribution
Used to describe probability where every event has equal chances to occur.
It has three parameters:
a - lower bound- default 0.0

b - upper bound- default 1.0

size - The shape of the returned array

```python
from numpy import random
x = random.uniform(size=(2,3))
print(x)
```

#### Visualization of Uniform Distribution

```python
sns.distplot(random.uniform(size = 1000), hist = False)
plt.show()
```

### Format Strings using fmt
### fmt syntax = marker|line|color

### Color Reference
'r' - Red

'g' - Green

'b' - Blue

'c' - Cyan

'm' - Magenta

'y' - Yellow

'k' - Black

'w' - White

### Line Reference
'-' - Solid Line

':' - Dotted line

'--' - Dashed line

'-.' - Dashed/dotted line

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,9,10,1])

plt.plot(ypoints, 'o:r') # o = circle : - dotted line r- red color
plt.show()

plt.plot(ypoints, 's-.b') # s = square marker, -. = line and dot b= blue color
plt.show()
```

### Marker Size :
#### argument markersize or 'ms' to define the size of the markers
#### argument markeredgecolor or 'mec' to set the color of the edge
#### argument markerfacecolor or 'mfc' to set the color inside the edge of the markers

```python
ypoints = np.array([5,2,9,11])

plt.plot(ypoints, marker = 'D', ms = 20) #marker diamond and markersize 20
plt.show()

plt.plot(ypoints, marker = '*', ms = 25, mec = 'm')
plt.show()

plt.plot(ypoints, marker = 's', ms = 20, mfc = 'r')
plt.show()

#use both mec and mfc to color the entire marker
plt.plot(ypoints, marker = 'H', ms = 25, mec = 'm', mfc = 'm')
plt.show()
```

### Matplotlib Line
#### Using argument 'linestyle' or 'ls' to change the style of line
#### to color the line argument 'color' or 'c' to color the line
#### using argument 'linewidth' or 'lw' we can change the width of line

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4,1,7,9,5])

plt.plot(ypoints, linestyle = '-.') # : = dotted, -- = dashed, -. = das n dot
plt.show()

plt.plot(ypoints, ls = ':', color = 'g', lw = '4.5') # : = dotted, -- = dashed, -. = dashdot
plt.show()
```

### ColorMap
##### The matplotlib module has a number of available colormaps.
##### Colormap is a list of colors from the range of 0 to 100
##### colormap is called "viridis" which is built in colormaps in matplotlib
##### colormap can be specified with "cmap" argument

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array([0,10,20,25,30,35,40,45,50,55,60,70,80])
plt.scatter(x,y, c = colors, cmap = 'viridis')
plt.colorbar()# to show the colorbar beside the plt
plt.show()
```

#### We can change the size of the dots

```python

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y, s = sizes)
plt.show()
```

#### By Alpha arguement we can adjust the transparency

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y, s = sizes, alpha = 0.5)
plt.show()
```

### Matplotlib Pie Charts

#### Creating pie chat using pie()
#### By default it starts from x-axis
#### and set counterclockwise

```python

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

#### labels of pie chart

```python
y = np.array([35,25,24,16])
mylabels = ["Apple","Orange","cheery","Banana"]

plt.pie(y, labels = mylabels)
plt.show()
```

### Start angle of pie chart
#### By default start angle is at x-axis
#### we can define it by startangle parameter

```python
y = np.array([35,25,24,16])
mylabels = ["Apple","Orange","cheery","Banana"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

### Explode 
#### explode parameter helps to wedges stand out


```python
y = np.array([35,25,24,16])
mylabels = ["Apple","Orange","cheery","Banana"]
myexplode = [0.2,0,0,0] #distance from the center

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode)
plt.show()
```

```python
#shadow parameter can create a shadow 
y = np.array([35,25,24,16])
mylabels = ["Apple","Orange","cheery","Banana"]
myexplode = [0.2,0,0,0] #distance from the center

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True)
plt.show()
```

### colors of wdges
#### color of each wedge can be defined with colors parameter

```python
y = np.array([35,25,24,16])
mylabels = ["Apple","Orange","cheery","Banana"]
mycolors = ["black","hotpink","magenta","b"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

### Legend() function
#### to add a list of explanation for each wedge

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```
